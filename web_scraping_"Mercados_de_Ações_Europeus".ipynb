```python
import pandas as pd
import requests
```

```python
#My user agent
header = {'User-Agente': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
```

```python

url = 'https://br.investing.com/equities/europe'
request = requests.get(url, header)

df = pd.read_html(request.text)
df
```

```python
for i , tabela in enumerate(df):
  print('----------------------------')
  print(i)
  print(tabela)
```
